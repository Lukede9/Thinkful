```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt

from matplotlib import rcParams
```

```python
%matplotlib inline

rcParams['figure.figsize'] = 7, 7

sns.set_style('whitegrid')
```

```python
countries = pd.read_csv("countries.csv")
```

Plot a variable in four different ways.

```python
#1 called distplot but seems like a histogram
sns.distplot(countries['Population'], bins=30)
```

```python
#2 Plotting just the population as a boxplot looked ugly so I added region
sns.boxenplot('Population', data=countries)
```

```python
#3
sns.boxplot('Population', data=countries)
```

```python
#4
sns.stripplot('Population', data=countries)
```

## Plot two continuous variables in three different ways.

```python
#1 Scatterplot of Population vs Area
sns.scatterplot('Area (sq. mi.)', 'Population', data=countries, hue='Region', \
               size=10)
```

```python
#2 I don't know what this is called or for
sns.lmplot(x='Area (sq. mi.)', y='Population', data=countries, \
          hue='Region')
```

```python
#3
sns.regplot('Area (sq. mi.)', 'Population', data=countries)
```

## Plot one continuous and one categorical variable six different ways.

```python
#1
sns.boxplot(countries['Population'], countries['Region'])
```

```python
#2
sns.catplot('Population', 'Region', kind="swarm", data=countries)
```

```python
#3
sns.violinplot('Region', 'Population', data=countries, hue='Region')
plt.xticks(rotation=45)
```

```python
#4
sns.stripplot('Population', 'Region', data=countries).set_title('Population by Region')
```

```python
#5
barplot = sns.barplot('Region', 'Population', data=countries).set_title('Population by Region')
plt.xticks(rotation=45,)
```

```python
#6
sns.pointplot('Population', 'Region', hue='Region', data=countries)
```
