```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('UNdata_Export_20180721_023440884.csv')
df
```

```python
plt.subplot(2, 2, 1)
plt.hist(df['Year'], bins=30)
plt.xlabel('Year')
plt.ylabel('Countries\nreporting in')

plt.subplot(2, 2, 2)
plt.plot(df['Count'])
plt.xlabel('Entry on list')
plt.ylabel('Homicides reported')

plt.subplot(2, 2, 3)
plt.hist(df['Rate'], bins=60)
plt.xlabel('Annual Homicides\nper 100,000')
plt.ylabel('# of countries')

plt.subplot(2, 2, 4)
plt.hist(df.Source.value_counts(), bins = 50)
plt.xlabel('Number of Reports\nby source')
plt.ylabel('Number of Sources')

plt.tight_layout()
plt.show()
```

The first histogram shows when the reports are from. From 1995 until 2008 more and more countries were reporting their homicide numbers or at least this data-set found more and more of them. This is significant because it means that even if the total number of murders increases, this might not be from more killing but from more reporting.

The second plot is a line plot and it does not really tell us much. Honestly, if you want to really stretch your imagination, it would be helpful if you wanted to scroll through the data and find the countries who had big reporting years. The first major blip in the data is about an eigth of the way through.

The first histogram on the second row shows that the vast majority of reporting countries have a rate of well below 50 annual homicides per 100,000 people. There are only a few countries even making it past the 25 point and it looks like 3 overachievers who knock off over 50 per.

The last histogram on the second row tells us how many reports came per source. Most of our sources just made a few reports and then one source is responsible for 400 reports. If there was one organization I wanted to know more about, it would be the one responsible for a third of the reports.
